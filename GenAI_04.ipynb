```python
def ANN(x1, x2, x3, w1, w2, w3, b):
    z = x1 * w1 + x2 * w2 + x3 * w3 + b
    return z
ANN(x1_train[0], x2_train[0], x3_train[0], w1, w2, w3, b)
```
